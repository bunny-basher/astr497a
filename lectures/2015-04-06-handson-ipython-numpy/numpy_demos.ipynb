```python
fig, ax = subplots(figsize=(12,8))
ax.scatter(stars['ra'], stars['dec'], marker='.', s=0.1)
```

```python
ra = stars['ra']
dec = stars['dec']
stars_subset = stars[(ra > 20) & (ra < 100) & (dec > 0)]

fig, ax = subplots(figsize=(12,8))
ax.scatter(stars_subset['ra'], stars_subset['dec'], marker='.', s=0.1)
```

```python
s = np.copy(stars)
ra = s['ra']

# rewrap to -180 + 180
#ra[ra > 180] = ra[ra > 180] - 360
ra[:] = (ra + 180) % 360 - 180

fig, ax = subplots(figsize=(12,8))
ax.scatter(s['ra'], s['dec'], marker='.')
```
